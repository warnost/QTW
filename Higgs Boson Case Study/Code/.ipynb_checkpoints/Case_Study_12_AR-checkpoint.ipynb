```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=100000, random_state=1776)
```

```python
# y = y_train.to_numpy()
# type(np.unique(y)[0])

y_train = y_train.astype(np.int64)
y_test = y_test.astype(np.int64)
```

```python
from tensorflow.keras import initializers

model = tf.keras.Sequential()
# input
model.add(tf.keras.Input(shape=(28,)))
# hidden
model.add(layers.Dense(300,
                       activation='relu', 
                       kernel_initializer=initializers.RandomNormal(stddev=0.1)))  # adds a layer with 300 neurons, tanh activation
model.add(layers.Dense(300,
                       activation='relu', 
                       kernel_initializer=initializers.RandomNormal(stddev=0.05)))
model.add(layers.Dense(300,
                       activation='relu', 
                       kernel_initializer=initializers.RandomNormal(stddev=0.05)))
model.add(layers.Dense(300,
                       activation='relu', 
                       kernel_initializer=initializers.RandomNormal(stddev=0.05)))
# output
model.add(layers.Dense(1,
                       activation='sigmoid',
                       kernel_initializer=initializers.RandomNormal(stddev=0.001)))  # adds a layer with 1 neurons, linear (aka NO) activation
```

```python
from tensorflow.keras import optimizers

opt = tf.keras.optimizers.SGD(learning_rate=0.05, momentum=1e-5)

model.compile(optimizer=opt,
              loss='binary_crossentropy',
              metrics=['accuracy','AUC'])
```
